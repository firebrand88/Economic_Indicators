```python
df_sp500 = pd.read_csv('./Inputs/SP500_data.csv')
df_sp500
```

```python
dict = {'Close*':'Close',
       'Adj Close**':'Adj_Close'}

df_sp500.rename(columns=dict,inplace=True)
```

```python
df_sp500.info()
```

```python
df_sp500['pct_dif_1'] = ''

for x in range(len(df_sp500.index)):
    if x > 0:
        df_sp500['pct_dif_1'][x] = ( df_sp500['Close'][x] - df_sp500['Close'][x-1] ) / df_sp500['Close'][x]
```
