```python
import pandas as pd

df = pd.read_csv('../Inputs/Inflation/bea_286_real_pce_formatted.csv')
df
```

```python
headers = list(df.columns)
```

```python
import re

new_headers = []

for header in headers:
#     print(header)
    x = re.sub(r'[^A-Za-z]','',header).lower()
    new_headers.append(x)
```
