```python
import pandas as pd

df = pd.read_csv('../Inputs/GDP/bea_116_real_gdp_formatted.csv')
df
```

```python
headers = list(df.columns)
headers
```

```python
import re

new_headers = []

for header in headers:
#     print(header)
    x = re.sub(r'[^A-Za-z]','',header).lower()
    new_headers.append(x)
```

```python
new_headers
```

```python
# Convert two lists to a dictionary
headers_dict = dict(zip(headers, new_headers))
headers_dict
```

```python
df2 = df.rename(columns=headers_dict)
df2
```

```python
df2.info()
```

```python
df3 = df2[['date','grossdomesticproduct']]
```
